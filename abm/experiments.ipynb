## imports

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

## basic plotting

```python
x = np.linspace(1, 10, 10)
y = x ** 2 - 1
plt.plot(x, y)
```

## figures

```python
figure = plt.figure(figsize=(5, 5))

y_2 = y * 1.2
y_3 = y * 1.6
y_4 = y * .6

ax = figure.add_axes([.1, .1, .9, .9])

ax.set_ylim(0, 100)

ax.grid(True, color="1")
ax.set_facecolor("0.95")

ax.set_ylabel("assesment")
ax.set_xlabel("time")

ax.plot(x, y, ".8", label="graph")
ax.plot(x, y_2, ".9", label="other graph")
ax.plot(x, y_3, ".2", label="whole other graph")
ax.plot(x, y_4, ".1")

ax.legend(loc=0)

```

## working with DataFrames

```python
data = pd.read_csv("agent_data.csv")
```

```python
time_x = np.linspace(0, 19, 20)

agent_y_values = []

for i in range(20):
    next_values = data[data["agent"] == i]["assesment"].values
    agent_y_values.append(next_values)
```

## plotting Hegselmann Krause agents

```python
figure = plt.figure(figsize=(5, 5))
ax = figure.add_axes([.1, .1, .9, .9])

ax.set_ylim(0, 1)
ax.set_xlim(0, 19)

ax.grid(True, color="1")
ax.set_facecolor("0.95")
ax.set_ylabel("assesment")
ax.set_xlabel("time")

for i in range(20):
    color = str(random.random() * .8)
    ax.plot(time_x, agent_y_values[i], color)

ax.legend(loc=0)
```

```python
figure = plt.figure(figsize=(5, 5))
ax = figure.add_axes([.1, .1, .9, .9])
ax.plot(x, y, "r", label="graph")

ax.clear()
ax.set_ylim(0, 1)
ax.set_xlim(0, 19)
ax.grid(True, color="1")
ax.set_facecolor("0.95")
ax.set_ylabel("assesment")
ax.set_xlabel("time")
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

def popup_bonus():
    win = tk.Toplevel()
    win.wm_title("Window")

    l = tk.Label(win, text="Input")
    l.grid(row=0, column=0)

    b = ttk.Button(win, text="Okay", command=win.destroy)
    b.grid(row=1, column=0)

def popup_showinfo():
    showinfo("Window", "Hello World!")

class Application(ttk.Frame):

    def __init__(self, master):
        ttk.Frame.__init__(self, master)
        self.pack()

        self.button_bonus = ttk.Button(self, text="Bonuses", command=popup_bonus)
        self.button_bonus.pack()

        self.button_showinfo = ttk.Button(self, text="Show Info", command=popup_showinfo)
        self.button_showinfo.pack()

root = tk.Tk()

app = Application(root)

root.mainloop()
```
